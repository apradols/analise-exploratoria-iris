# An√°lise explorat√≥ria do conjunto de dados 'Iris Species'

  Nesse projeto, foi utilizado o dataset do Kaggle [Iris Species](https://www.kaggle.com/datasets/uciml/iris), para o fim de praticar alguns conceitos de data science.

üìö Dicion√°rio dos dados

 * `Id`: N√∫mero de itendtifica√ß√£o da iris
 * `SepalLengthCm`: Comprimento da S√©pala em centimetros

 * `Species`: Esp√©cies

‚¨á Importando dataset

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/Iris.csv')
```

üåê Traduzindo colunas

```python
mapa = {
    'Id': 'id',
    'SepalLengthCm': 'comprimento_sepala',
    'SepalWidthCm': 'largura_sepala',
    'PetalLengthCm': 'comprimento_petala',
    'PetalWidthCm': 'largura_petala',
    'Species': 'especies'
}
dados = dados.rename(columns=mapa)
```

üìä Visualizando dados

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.countplot(x='especies', data=dados)
ax.set_title('Quantidade de cada esp√©cie de iris', fontsize=16)
ax.set_xlabel('Esp√©cies', fontsize=12)
ax.set_ylabel('Quantidade de dados', fontsize=12)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax = sns.scatterplot(data=dados, x='comprimento_petala', y='largura_petala', hue='especies')
ax.set_title('Compara√ß√£o de comprimento de p√©talas e largura das p√©palas em rela√ß√£o as esp√©cies', fontsize=15)
ax.set_xlabel('Comprimento de p√©talas', fontsize=12)
ax.set_ylabel('Lagura de p√©talas', fontsize=12)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax = sns.scatterplot(data=dados, x='comprimento_sepala', y='largura_sepala', hue='especies')
ax.set_title('Compara√ß√£o de comprimento de s√©pala e largura das s√©pala em rela√ß√£o as esp√©cies', fontsize=15)
ax.set_xlabel('Comprimento de s√©palas', fontsize=12)
ax.set_ylabel('Lagura de s√©palas', fontsize=12)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax = sns.scatterplot(data=dados, x='comprimento_sepala', y='largura_petala', hue='especies')
ax.set_title('Compara√ß√£o de comprimento de s√©pala e largura das p√©tala em rela√ß√£o as esp√©cies', fontsize=15)
ax.set_xlabel('Comprimento de s√©palas', fontsize=12)
ax.set_ylabel('Lagura de p√©talas', fontsize=12)
plt.show()
```

Boxplot

```python
def graph(y):
    sns.boxplot(x="especies", y=y, data=dados)

plt.figure(figsize=(12,12))

plt.subplot(221)
graph('largura_sepala')

plt.subplot(222)
graph('largura_petala')

plt.subplot(223)
graph('comprimento_sepala')

plt.subplot(224)
graph('comprimento_petala')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax = sns.heatmap(data=dados.corr(), cmap='rocket', annot=True)
ax.set_title('Gr√°fico de Correla√ß√£o')
plt.show()
```
